```python
data = pd.read_csv(max_file)
data
```

```python
sd=[]
for i in range(len(data['Name'])):
    x = float(data['Low'].iloc[i].replace(',',''))
    x = float("{0:.4f}".format(x))
    y = float(data['High'].iloc[i].replace(',',''))
    y = float("{0:.4f}".format(y))
    z = float(data['Last'].iloc[i].replace(',',''))
    z = float("{0:.4f}".format(z))
    print (x,y,z)
    
    if z >= x and z <= y :
        p = ((z-x)/(y-x))*100
    else:
        p = "Out of Days Range"
    sd.append(p)
    #if int(z) not in range(int(x), int(y),):
    #    sd.append('Out of Days Range')
    #else:
    #    sd.append(str(x)+"_"+str(y)+"_"+str(z)+'_In Days Range')
print(sd)

data['Day Range']=sd
data['Day Range']
```

```python
#'15Mins to Daily'
bs1= []
for i in range(len(df['15 Minutes'])):
    if "-"not in df['Chg. %'][i] and df['15 Minutes'][i] > 0 and df['30 Minutes'][i] > 0 and df['Hourly'][i] > 0 and df['5 Hours'][i] > 0 and df['Daily'][i] > 0:
        if df['15 Minutes'][i] == df['30 Minutes'][i] == df['Hourly'][i] == df['5 Hours'][i] ==df['Daily'][i] == 2:
            bs1.append('1_Strong Buy')
        else:
            x = df['15 Minutes'][i]+df['30 Minutes'][i]+df['Hourly'][i]+df['5 Hours'][i]+df['Daily'][i]
            bs1.append('2_Buy:'+str(x))
    
    elif "-" in df['Chg. %'][i] and df['15 Minutes'][i] < 0 and df['30 Minutes'][i] < 0 and df['Hourly'][i] < 0 and df['5 Hours'][i] < 0 and df['Daily'][i] < 0:
        if df['15 Minutes'][i] == df['30 Minutes'][i] == df['Hourly'][i] == df['5 Hours'][i] ==df['Daily'][i] == -2:
            bs1.append('1_Strong Sell')
        else:
            x = df['15 Minutes'][i]+df['30 Minutes'][i]+df['Hourly'][i]+df['5 Hours'][i]+df['Daily'][i]
            bs1.append('2_Sell:'+ str(x))
    else:
        bs1.append('0')

df['Buy_Sell']=bs1
df
```

```python
for i in range(len(df['Buy_Sell'])):
    if df['Buy_Sell'][i]==str(0):
        df = df.drop(i)
df.reset_index(inplace = True)
df
```
